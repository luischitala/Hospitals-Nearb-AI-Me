```python
def quitar_articulos(texto):
    tokens = texto.split(' ')

    articulos = ['EL','LA','LOS','LAS']
    for i,t in enumerate(tokens):
        if t in articulos:
            del tokens[i]
            
    tokens = " ".join(tokens)
    return tokens

    
```

```python
df2['Nombre de la Unidad'].apply(lambda x: quitar_articulos(x)).head(20)

```
